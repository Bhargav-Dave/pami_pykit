```python
from jpmesh import parse_mesh_code
```

```python
val=[]
input1='fin_op#2'
fout='fin_op#3'
lines=[]
with open(input1,'r') as k:
    for line in k:
        lines.append(line)
        li=line.split()
        val.append(li)
```

```python
# outfile=sys.argv[2]
kq=1
with open(fout, 'w') as f:
    for i in range(len(lines)):
        for m in val[i]:
            mesh = parse_mesh_code(m)
            mesh_center = mesh.south_west + (mesh.size / 2.0)
            x=(mesh_center.lon.degree,mesh_center.lat.degree)
            n_w=str((str(mesh.south_west.lon.degree)+' '+str(mesh.south_west.lat.degree),str(mesh.south_west.lon.degree)+' '+str(mesh.south_west.lat.degree+mesh.size.lat.degree),str(mesh.south_west.lon.degree+mesh.size.lon.degree)+' '+str(mesh.south_west.lat.degree+mesh.size.lat.degree),str(mesh.south_west.lon.degree+mesh.size.lon.degree)+' '+str(mesh.south_west.lat.degree),str(mesh.south_west.lon.degree)+' '+str(mesh.south_west.lat.degree)))
#             borders=(s_w,s_e,n_e,n_w,s_w)
#             print(borders)
#             print(lines[i].rstrip())
            n_w.replace("'","")
            n_w='POLYGON('+n_w+')'
            print(n_w)
            f.write('%s'%m.rstrip())
            f.write('\t%s'%kq)
            f.write('\t%s'%str(n_w))
#             f.write(' %s'%str(x))
            f.write('\n')
        kq+=1
    f.close()
```
